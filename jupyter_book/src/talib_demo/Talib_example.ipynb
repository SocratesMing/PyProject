```python
import talib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1 = np.random.randint(500, size = (100,5))
data = pd.DataFrame(data1, columns= ["close","high", "low","open", "volume"])
print(data.head())
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
close = data['close']
```

```python
# 一、重叠研究（overlap studies）
# 1.简单移动平均指标SMA
# 参数说明：talib.SMA(a,b)
# a:要计算平均数的序列；b:计算平均线的周期。表示计算a的b日移动平均
close = data['close']
SMA = talib.SMA(close,5)
x = SMA
plt.plot(SMA, color = "b")
plt.title("SMA")
plt.show()
```

```python
# 2.布林线BBANDS
# 参数说明：talib.BBANDS(close, timeperiod, matype)
# close:收盘价；timeperiod:周期；matype:平均方法(bolling线的middle线 = MA，用于设定哪种类型的MA)
# MA_Type: 0=SMA, 1=EMA, 2=WMA, 3=DEMA, 4=TEMA, 5=TRIMA, 6=KAMA, 7=MAMA, 8=T3 (Default=SMA)
upper, middle, lower = talib.BBANDS(close,5,matype = talib.MA_Type.EMA)
plt.plot(upper, color = "g", label = "upper")
plt.plot(middle, color = "b", label = "middle")
plt.plot(lower, color = "r", label = "lower")
plt.legend()
plt.title("BBANDS")
plt.show()
```

```python
# print("upper, middle, lower", upper, middle, lower)
# 3. DEMA 双移动平均线:DEMA = 2*EMA-EMA(EMA)
# 参数说明：talib.DEMA(close, timeperiod = 30)
DEMA = talib.DEMA(close, timeperiod = 3)
plt.plot(DEMA, color = "b")
plt.title("DEMA")
plt.show()

```

```python
# 4. MA
# 参数说明：MA(close, timeperiod = 30, matype=0)
# close:收盘价；timeperiod:周期；matype:计算平均线方法
MA = talib.MA(close, timeperiod = 10, matype = 0)
plt.plot(MA, color = "b")
plt.title("MA")
plt.show()
```

```python
# 5. EMA
# 参数说明：EMA = talib.EMA(np.array(close), timeperiod=6)
# close:收盘价；timeperiod:周期；matype:计算平均线方法
EMA = talib.EMA(close, timeperiod = 6)
plt.plot(EMA, color = "b")
plt.title("EMA")
plt.show()
```

```python
# 6.KAMA：考夫曼的自适应移动平均线
# 参数说明：KAMA = talib.KAMA(close, timeperiod = 30)
KAMA = talib.KAMA(close, timeperiod = 30)
plt.plot(KAMA, color = "b")
plt.title("KAMA")
plt.show()
```

```python
# 7. MIDPRICE：阶段中点价格
# talib.MIDPOINT(close, timeperiod)
# 参数说明：close:收盘价；timeperiod:周期；
MIDPOINT = talib.MIDPOINT(close, timeperiod=14)
plt.plot(MIDPOINT, color = "b")
plt.title("MIDPOINT")
plt.show()
```

```python
# 8.SAR：抛物线指标
# SAR(high, low, acceleration=0, maximum=0)
# 参数说明：high：最高价；low:最低价；acceleration：加速因子；maximum：极点价
SAR = talib.SAR(data['high'], data['low'], acceleration=0, maximum=0)
plt.plot(SAR, color = "b")
plt.title("SAR")
plt.show()
```

```python
# 9.MIDPRICE：阶段中点价格（Midpoint Price over period）
# talib.MIDPOINT(close, timeperiod=14)
# 参数说明：close:收盘价；timeperiod:周期；
MIDPOINT = talib.MIDPOINT(close, timeperiod=14)
plt.plot(MIDPOINT, color = "b")
plt.title("MIDPOINT")
plt.show()
```
