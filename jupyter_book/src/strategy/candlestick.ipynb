```python
import plotly.graph_objects as go

import pandas as pd
from datetime import datetime

df = pd.read_csv(r'E:\Project\PyProject\JupyterBook\strategy\apple.csv')

fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['AAPL.Open'],
                high=df['AAPL.High'],
                low=df['AAPL.Low'],
                close=df['AAPL.Close'])])

fig.show()
```
