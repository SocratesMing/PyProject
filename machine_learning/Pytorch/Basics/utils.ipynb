# unsqueeze

是 PyTorch 中用于扩展张量维度的核心函数，通过在指定位置插入尺寸为 1 的新维度来调整张量形状，常用于数据广播、模型输入适配等场景。以下是其详细解析：

```python
import torch
from torch import nn
x = torch.tensor([1, 2, 3, 4])  # shape: [4]
x0 = x.unsqueeze(0)  # tensor([[1, 2, 3, 4]])
print(x0, x0.size())
x1 = x.unsqueeze(1)  # tensor([[1], [2], [3], [4]])
print(x1, x1.size())
```

```python
y = torch.randn(2, 3)  # shape: [2, 3]
x0 = y.unsqueeze(0)
print(x0, x0.size())
x1 = y.unsqueeze(1)
print(x1, x1.size())
x2 = y.unsqueeze(2)
print(x2, x1.size())
```

# Flatten
将张量展开



```python
input = torch.randn(32, 1, 5, 5)
# With default parameters 默认从dim1到dim last
m = nn.Flatten()
output = m(input)
print(output.size())
# With non-default parameters
m = nn.Flatten(0, 2)
output = m(input)
output.size()
```
